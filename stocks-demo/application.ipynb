```python
import datetime
import requests
import os 

def ingestion_service_stocks(endpoint): 
    now = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    t = requests.post(endpoint,json={'ticker':['stocks'],
                                     'Datetime': now,
                                     'start_delta':3,
                                     'end_delta':0,
                                     'interval':'5m',
                                     'n_stocks':4})
    
    print(t.text)
    
def ingestion_service_news(endpoint): 
    now = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    t = requests.post(endpoint,json={'ticker':['news'],
                                     'Datetime': now,
                                     'n_stocks':4})
    print(t.text)
    

username = os.environ['V3IO_USERNAME']
news_internal = f'http://stocks-{username}-get-stocks-stocks-{username}.default-tenant.app.vmdev94.lab.iguazeng.com/'
stocks_internal = f'http://stocks-{username}-get-news-stocks-{username}.default-tenant.app.vmdev94.lab.iguazeng.com/'

ingestion_service_stocks(stocks_internal)
ingestion_service_news(news_internal)
```
